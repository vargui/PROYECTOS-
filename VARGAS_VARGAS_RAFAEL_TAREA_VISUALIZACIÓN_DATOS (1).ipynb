AHORA GENERAMOS UN PEQUE√ëO RECOMENDADOR DE CANCIONES SEG√öN GUSTO DEL USUARIO. (LOS NOMBRES DE ARTISTAS Y GENEROS QUE SE PUEDEN SOLICITAR TIENEN QUE ESTAR DENTRO DEL DATASET EN EL TREEMAP DE ARRIBA)

```python
def recomendar_canciones_por_genero_artista(df, genero_usuario, artista_usuario):

    canciones_genero = df[df['genre'].str.contains(genero_usuario, case=False, na=False)]


    canciones_filtradas = canciones_genero[~canciones_genero['artist'].str.contains(artista_usuario, case=False, na=False)]


    if canciones_filtradas.empty:
        print(f"No se encontraron canciones de otros artistas en el g√©nero: {genero_usuario}")
        return


    recomendaciones = canciones_filtradas.sample(n=min(3, len(canciones_filtradas)))


    print(f"\nüéµ Basado en que te gusta {artista_usuario} ({genero_usuario}), te recomiendo:\n")
    for idx, row in recomendaciones.iterrows():
        print(f"- {row['song']} de {row['artist']}")


genero_favorito = input("¬øCu√°l es tu g√©nero favorito?  ")
artista_favorito = input("¬øCu√°l es tu artista favorito?  ")
recomendar_canciones_por_genero_artista(df, genero_favorito, artista_favorito)

```
